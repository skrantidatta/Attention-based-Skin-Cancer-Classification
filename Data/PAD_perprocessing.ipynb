```python
import pandas as pd
import numpy as np
import os
```

```python
data_pd = pd.read_csv('D:/AI_CAPSTONE/skin_lesion/Dataset/PAD-UFES-20/PAD-UFES-20_metadata.csv')
data_pd.head()
```

```python
df_count = data_pd.groupby('lesion_id').count()
df_count.head()
```

```python
df_count = df_count[df_count['diagnostic'] == 1]
df_count.reset_index(inplace=True)
```

```python
def duplicates(x):
    unique = set(df_count['lesion_id'])
    if x in unique:
        return 'no' 
    else:
        return 'duplicates'
```

```python
data_pd['is_duplicate'] = data_pd['lesion_id'].apply(duplicates)
data_pd.head()
```

```python
df_count = data_pd[data_pd['is_duplicate'] == 'no']
```

```python
df_count.head()
```

```python
def splt(x):
    x.strip().split('.png')
    return x
```

```python
apply = df_count['img_id']
new_img_id = apply.apply(splt)
df_count['img_id'] = new_img_id
```
